1. Fórmulas de ConversiónPara pasar de una escala a otra, utilizamos las siguientes relaciones matemáticas:
* De Celsius a Fahrenheit: Multiplicamos por $1.8$ (o $9/5$) y sumamos $32$.$$°F = (°C \times 1.8) + 32$$
* De Fahrenheit a Celsius: Restamos $32$ y luego dividimos por $1.8$ (o multiplicamos por $5/9$).$$°C = \frac{°F - 32}{1.8}$$
2. Ejemplos Resueltos Paso a PasoCaso A: Convertir $25°C$ a $°F$ (Temperatura ambiente)
* Identificar datos: Tenemos $25°C$.
* Sustituir en la fórmula: $°F = (25 \times 1.8) + 32$.
* Multiplicar: $25 \times 1.8 = 45$.
* Sumar: $45 + 32 = 77$.
* Resultado: $25°C$ equivale a $77°F$.
* Caso B: Convertir $98.6°F$ a $°C$ (Temperatura corporal promedio)
* Identificar datos: Tenemos $98.6°F$.
* Sustituir en la fórmula: $°C = (98.6 - 32) / 1.8$.
* Restar: $98.6 - 32 = 66.6$.
* Dividir: $66.6 / 1.8 = 37$.
* Resultado: $98.6°F$ equivale a $37°C$

```python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"---> {celsius}°C equivalen a {fahrenheit:.2f}°F")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"---> {fahrenheit}°F equivalen a {celsius:.2f}°C")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Iram Jimenez NL = 28 ")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Llamada inicial para arrancar el programa
if __name__ == "__main__":
    menu()
```
